## Simulate circuits using the 'quantum_info' module

```python
from qiskit import QuantumCircuit
import qiskit.quantum_info as qi
```

```python
qc=QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.draw('mpl')
```

```python
psi=qi.Statevector(qc)   #Statevector is a class that we ae calling from quantum_info module
psi.draw('latex')
```

**Note**: The state is in superposition of standard basis states (equally probable)

We may also find the probabilities of each of the possible states in the following way.

```python
probs=psi.probabilities()
print(probs)
```

<br>

We may simulate the state preparing process using the *sample_counts* for a given number of shots.

```python
counts=psi.sample_counts(shots=1024)
print(counts)
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

<br>

We may visualize this using the *plot_histogram* package,

We may also check which state appears in which shot using the *sample_memory* function.

```python
samples=psi.sample_memory(shots=10)
print(samples)
```

<br>

We may now evaluate the expectation value of an operator for this particular state using the **SparsePauliOp**

```python
o= qi.SparsePauliOp(['ZZ','XX','II'],[0.2,-0.3,0.4])  # creating the operator
oo=o.to_operator()  # writing it in the matrix form
oo.draw('latex')
```

```python
o_exp=psi.expectation_value(o)
print(o_exp)
```

```python
psi.draw('bloch') # representing the state in Bloch sphere
```

<br>

We may also check the 'density matrix' for this particular state.

```python
rho= qi.DensityMatrix(qc)
rho.draw('latex')
```

```python
rho.draw('city')   # city-scale map for the density matrix
```

```python
rho1=qi.partial_trace(rho,[0])
rho1.draw('latex')
```

<br>

A quantum state evolution is guided by the <u>Hamiltonian unitary</u>. We may have an idea looking at the corresponding unitary.

## Simulate circuit using *BasicSimulator*

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.providers.basic_provider import BasicSimulator

simulator_basic= BasicSimulator()

X= QuantumRegister(1,"q1")
Y= QuantumRegister(1,"q0")
A= ClassicalRegister(2,"c")
circuit= QuantumCircuit(X,Y,A)

circuit.h(X)
circuit.measure(X,A[1])
circuit.x(Y).c_if(A,1)
circuit.measure(Y,A[0])

circuit.draw('mpl')
```

```python
job= simulator_basic.run(circuit,shots=1024)
result= job.result()
print(result)
```

```python
counts= result.get_counts()
print(counts)
```

```python
from qiskit.visualization import plot_histogram, plot_distribution
plot_distribution(counts)
```

```python
job = simulator_basic.run(circuit, shots=10, memory=True)
result= job.result()
samples= result.get_memory()
print(samples)
```

<br>

This is for the constructed circuit.
Now we will try to simulate the QFT circuit.

```python
from qiskit.circuit.library import QFT

qc= QuantumCircuit(3,3)    # 3,3 stands for 3 quantum and 3 classical bits
qc.append(QFT(3),range(3))
qc.measure(range(3), range(3))

qc.draw('mpl') 
```

<br>

If we go to simulate this circuit error pops up, because the QFT is not a part of the basis gates.

So here we are to use *transpiler* that will turn this into a linear combination of basis gates that backend will understand.

```python
from qiskit import transpile

qc_t= transpile(qc, backend=simulator_basic)

job=simulator_basic.run(qc_t, shots= 1024)
```

```python
result= job.result()
counts= result.get_counts()
print(counts)
```

Now what are **Primitives**?

Primitives serve as an abstraction between user and the backend. It is a basic functional block to run slightly higher-level quantum programs and return more than just <u>counts</u> at the output.

There are two types of '<u>primitive</u>': sampler primitive and estimator primitive.<br>
<u>Sampler primitive</u> sample output state of a circuit.<br>
<u>Estimator primitive</u> estimate expectation value of operators in a circuit.

**Note:** Primitives take as inputs more than just circuits. They take PUBs (Primitive Unified Blocs).

We must carefully note that 'primitive' belongs to *qiskit* package as well as *qiskit_ibm_runtime* package, but the workflow of them are different.

<br>

<u>V1 Primitives</u>

In Qiskit:<br>
*Sampler*, *Estimator*

In Qiskit Aer:<br>
*Sampler*, *Estimator*

In Qiskit IBM Runtime:<br>
*Sampler(V1)*, *Estimator(V2)*

## <u>Optimize for hardware - techniques</u> **:**


*transpile* your abstract circuit into a circuit that can run on target hardware.

Now the target hardware is a Real quantum device and they are subject to constraints:<br>
- **Basis gate set**: Only a limited set of gates can be executed directly on the hardware. Other gates must be rewritten in terms of these basis gates. <br>
- **Qubit connectivity**: Only certain pairs of qubits acn be directly interacted with each other.<br>
- **Errors**: Each operation has a chance of *error*, so circuit optimizations can greatly affect performance.

We can also *verify* our circuit with simulation using any of these simulation tools: Qiskit SDK reference primitives, Qiskit Runtime local testing and Qiskit Aer. 

## <u>Execute on target hardware</u> **:**

The inputs to a Primitive is a PUB (**Primitive Unified Blocs**).

We may execute the operation using Qiskit IBM Runtime. There are three types of  *mode* for that -- <br>
1. Single job <br>
2. Batch: multiple concurrent jobs <br>

## Why do we need quantum circuit compilation (transpilation)?

- Despite programming at a low level there is still abstractions in our representation that can't be executed by hardware.<br>
- There are numerous contraints with hardware that need to be accounted for.<br>
- The transpiler's job is to take the high-level abstract circuit and output an optimized circuit that is capable of running on a given quantum computer.

## Modelling Hardware constraints

**Operations**<br>
Each Quantum Computer only supports a set of operations. Typically, hardware supports a universal gate set that can be used to construct any unitary operation.<br>
All qubits do not necessarily need to support the same set of operations.

**Qubit connectivity**<br>
- Qubits may have limited connectivity<br>
- For IBM's super conducting hardware this is limited to qubits that have physical connectivity<br>
- For multi-qubit gates this means we can only run these operations where there is connectivity between qubits

Quantum circuits are compiled (by transpiler) and run physically on the quantum hadware; the performance and power of today's quantum processors are still largely limited by noise.<br> We try to understand the limitation of quantum hardware due to noise so that we can make best use of it.

## What are the sources of noise?

1. **State Preparation And Measurement (SPAM) erors**<br>
2. **Coherent errors in gates:** Incorrect Hamiltonian evolution<br>
3. **Incoherent errors in gates:** Loss of quantum information in the form of superposition and entanglement <br>

**We must note :**<br>
- Coherent errors often result in oscillation in signals
- It can build up much more rapidly than incoherent errors
- A number of solutions for coherent errors: *dynamical decoupling*, *Twirling*

## Quantum state and effect of noise

<img src="nc.png" style="width:387px; height:230px"/>

The density matrix representation of a pure state is $\rho$= $\ket{\psi}\bra{\psi}$ and that of a mixed state is $\rho$=$\Sigma q_{i}\ket{\psi_{i}}\bra{\psi_{i}}$<br>
What noise does is it turns a pure state into a mixed state via a noise channel.
